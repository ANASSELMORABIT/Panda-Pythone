```python
#Eliminar del dataframe les fileres que tinguin tots els seus camps a valor nul. Els canvis 
# han de ser permanents en el dataframe avocados.
dataframe.dropna(how='any',inplace=True)
```

```python
# Eliminar del dataframe les fileres duplicades. Els canvis han de ser permanents en el 
# dataframe avocados. 
dataframe.drop_duplicates(inplace=True)
```

```python
# r) Eliminar del dataframe les fileres que corresponguin a la filera d’encapçalament 
# repetida. 
dataframe.drop_duplicates(keep='first')
```

```python
# s) Eliminar del dataframe les columnes Unnamed: 0. 
dataframe.drop('Unnamed: 0',axis=1,inplace=True)
```

```python

# Reorganitzar l’ordre de les columnes del dataframe de la següent manera: ['Date', 
# 'year', 'AveragePrice', 'Total Volume', '4046', '4225', '4770', 'Total Bags', 'Small Bags', 
# 'Large Bags', 'XLarge Bags','region'] 
dataframe=dataframe[['Date', 'year', 'AveragePrice', 'Total Volume', '4046', '4225', '4770', 'Total Bags', 'Small Bags', 'Large Bags', 'XLarge Bags','type','region']]
```

```python
# Convertir la columna AveragePrice a tipus numèric. Els canvis han de ser permanents 
# en el dataframe avocados. 
dataframe['AveragePrice'] = dataframe['AveragePrice'].astype('int64')

```

```python
# Convertir la columna Total Volume a tipus numèric. Els canvis han de ser permanents 
# en el dataframe avocados. 
dataframe['Total Volume']=dataframe['Total Volume'].astype('int64')
```

```python
# w) Convertir les columnes Small Bags, Large Bags i XLarge Bags a tipus numéric. 
dataframe.astype({'Small Bags': 'float32', 'Large Bags': 'float32', 'XLarge Bags': 'float32'}).dtypes
```

```python
# x) Convertir la columna Date a tipus de data. Els canvis han de ser permanents en el 
# dataframe avocados. 
dataframe['Date']=dataframe['Date'].astype('datetime64[ns]')
```

```python
# y) Crear una llista total_sales_per_size, que contingui la suma total de les columnes 
# Small Bags, Large Bags i XLarge Bags, respectivament. 

total_sales_per_size = [dataframe['Small Bags'].sum(),dataframe['Large Bags'].sum(),dataframe['XLarge Bags'].sum()]
total_sales_per_size
```

```python
#z) Fer un gràfic de tarta que representi els valors de total_sales_per_size. 
from matplotlib import pyplot as plt
fig,ax = plt.subplots(figsize=(15,8))
x=['Small Bags','Large Bags','XLarge Bags']
y=[3324870837.51, 991615770.5500001, 56689177.32999999]
ax.plot(x,y)
plt.show()
```

```python
# aa) Crear una llista regions que contingui els valors que apareixen a la columna region del 
# dataframe avocados, sense repeticions. 
regions = dataframe['region'].unique()
regions
```

```python
# bb) Crear en el dataframe avocados una nova columna month a partir de la columna Date. 
dataframe['month']=dataframe['Date'].dt.month
```

```python
# cc) Crear en el dataframe avocados una nova columna Total Price a partir de les columnes 
# Average Price i Total volume. 
dataframe['Total Price']=dataframe['AveragePrice']*dataframe['Total Volume']
```

```python
# dd) Modificar la columna type para que sus valores sean sólo cono y org. Usando 
# el método .str[slice].
dataframe['type'] = dataframe['type'].str[:2]
dataframe['type'].unique()
```
